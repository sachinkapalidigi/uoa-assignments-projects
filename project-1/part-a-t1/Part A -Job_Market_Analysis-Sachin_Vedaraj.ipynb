```python
# Searching Data Scientist on Indeed. Scraping data using selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
import os
import time

# https://sites.google.com/chromium.org/driver/

class IndeedScraper:
    def __init__(self):
        self.base_url = "https://in.indeed.com/jobs"
        self.service = Service(executable_path="./chromedriver")
        options = Options()
        # options.headless = True
        self.driver = webdriver.Chrome(service=self.service, options=options)

    def buildUrl(self, query_params={}):
        """Constructs the URL with optional query parameters."""
        url = self.base_url
        if query_params:
            query_string = "&".join([f"{key}={value}" for key, value in query_params.items()])
            url += "?" + query_string
        return url

    def fetch(self, query_params={}):
        """Makes the HTTP request and returns the content."""
        url = self.buildUrl(query_params)
        page_source = None
        try:
            self.driver.get(url)
            time.sleep(5)
            page_source = self.driver.page_source
        finally:
            print("Fetching completed")
        return page_source
    
    def extract_job_info(self):
      job_listings = self.driver.find_elements(By.CSS_SELECTOR, "#mosaic-provider-jobcards li")
      jobs = []

      for listing in job_listings:
          try:
              title = listing.find_element(By.CSS_SELECTOR, ".jobTitle > a").find_element(By.CSS_SELECTOR, "span").text
          except:
              title = None
          
          try:
              company = listing.find_element(By.CSS_SELECTOR, "span[data-testid='company-name']").text
          except:
              company = None

          try:
              location = listing.find_element(By.CSS_SELECTOR, "div[data-testid='text-location']").text
          except:
              try:
                location = listing.find_element(By.CSS_SELECTOR, "span[data-testid='text-location']").text
              except:
                location = "N/A"

          try:
              rating = listing.find_element(By.CSS_SELECTOR, "span[data-testid='holistic-rating']").text
          except:
              rating = "N/A"

          try:
              company_metadata_div = listing.find_element(By.CSS_SELECTOR, ".jobMetaDataGroup")
          except:
              company_metadata_div = None

          try:
              salary = company_metadata_div.find_element(By.CSS_SELECTOR, ".salary-snippet-container").text
          except:
              salary = "N/A"

          try:
              if "Full-time" in company_metadata_div.text:
                  job_type = "Full-time"
              elif "Part-time" in company_metadata_div.text:
                  job_type = "Part-time"
              else:
                  job_type = "N/A"
          except:
              job_type = "N/A"

          if title != None and company != None:
            jobs.append({
              'title': title,
              'company': company,
              'location': location,
              'rating': rating,
              'salary': salary,
              'job_type': job_type,
            })

      return jobs


# Usage
scraper = IndeedScraper()
jobs = []

for page in range(11, 50):
    start_param = page * 10  # Page 1 starts at 0, Page 2 starts at 10, and so on.
    query_params = {'q': 'data scientist', 'l': 'India'}
    if page > 0:  # From the second page onwards, add the 'start' parameter
        query_params['start'] = start_param
    scraper.fetch(query_params)
    page_jobs = scraper.extract_job_info()
    jobs.extend(page_jobs)

# Convert the list of job dictionaries to a DataFrame
jobs_df = pd.DataFrame(jobs)

# Check if the file exists
file_exists = os.path.isfile('indeed_jobs.csv')

# Write to CSV, appending if file exists, with no header if appending
jobs_df.to_csv('indeed_jobs.csv', mode='a', header=not file_exists, index=False)

```
